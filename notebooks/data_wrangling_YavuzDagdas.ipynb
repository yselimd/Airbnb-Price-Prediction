Looks like there are several things I need to do to clean and prepare the data:
1. Change the data type to datetime: last_scraped, host_since, calendar_last_scraped,first_review,last_review
2. Check following columns for unique responses (I see t or f assuming that they represent true or false) and convert them to 0,1: host_is_superhost,host_has_profile_pic,host_identity_verified,has_availiability,instant_bookable 
3. Price data is entered as "$100" and is recognized as object. I need to remove the dollar sign and convert data type to numeric. 

```python
# changing data type of is,host_id columns to object
listings_detail[['id','host_id']] = listings_detail[['id','host_id']].astype('object')
```

```python
# Changing data type to datetime for these columns.

listings_detail[['last_scraped', 'host_since', 'calendar_last_scraped','first_review','last_review']] = listings_detail[['last_scraped', 'host_since', 'calendar_last_scraped','first_review','last_review']].apply(pd.to_datetime)
```

```python
# Checking the unique values in following columns: 'host_is_superhost','host_has_profile_pic','host_identity_verified','has_availability','instant_bookable'

ls = ['host_is_superhost','host_has_profile_pic','host_identity_verified','has_availability','instant_bookable']
[print(i,' has following unique values: ',listings_detail[i].unique()) for i in ls];
```

There are 3 unique values in these columns: 't','f' and nan. I will convert these values to 1 and 0 , respectively.

```python
# Converting t and f to 1 and 0.

listings_detail[['host_is_superhost','host_has_profile_pic','host_identity_verified','has_availability','instant_bookable']] = listings_detail[['host_is_superhost','host_has_profile_pic','host_identity_verified','has_availability','instant_bookable']].replace({'t':1,'f':0})
[print(i,' has following unique values: ',listings_detail[i].unique()) for i in ls]
```

```python
# Price column type is object. Contains $ sign and comma in some rows, 
# so I will remove both and convert the remaining string to numeric.

listings_detail['price'] = listings_detail['price'].str.replace(',','').str.replace('$','')
listings_detail['price'] = pd.to_numeric(listings_detail['price'])
```

```python
# Creating a list that has the column name of columns that do not contain relevant or 
# useful information or completely null.

columns_to_drop = ['listing_url', 'scrape_id', 'last_scraped', 'name', 'description',
       'neighborhood_overview', 'picture_url', 'host_url',
       'host_name', 'host_location', 'host_about', 'host_thumbnail_url', 'host_picture_url',
       'host_neighbourhood', 'neighbourhood_group_cleansed','bathrooms','calendar_updated']

# Dropping columns in the list.
listings_detail = listings_detail.drop(columns_to_drop,axis=1)
```

```python
# Bathrooms column was completely null but there is another column 'bathrooms_text'
# Let's look at the unique values in the 'bathroom_text' column.

listings_detail['bathrooms_text'].value_counts()
```

* I will convert these strings to numeric values but how can I decide what to do with private bath shared bath?
* Could these be specific to certain type of listings; i.e. home, hotel, private room etc?

```python
room_type = listings_detail.room_type.unique()
[print(i, ' has following bathrooms:\n', listings_detail[listings_detail['room_type']==i]['bathrooms_text'].unique(),'\n') for i in room_type];
```

* So private or shared bathrooms are specific for listings that are Private room or Shared room

```python
# Replacing '...half-bathooms' with 0.5 and removing the letters from the remaining rows leaving only numbers.
listings_detail['bathrooms_text'] = listings_detail['bathrooms_text'].replace('Half-bath','0.5').replace('Shared half-bath','0.5').replace('Private half-bath','0.5').str.replace(' shared baths','').str.replace(' shared bath','').str.replace(' private bath','').str.replace(' baths','').str.replace(' bath','')
```

```python
# Changing data type of bathrooms_text column to numeric.
listings_detail['bathrooms_text'] = pd.to_numeric(listings_detail['bathrooms_text'])
```

```python
# Renaming bathrooms_text column to bathrooms.
listings_detail.rename(columns={'bathrooms_text':'bathrooms'},inplace=True)
```

```python
# checking the unique values for the host response rate since it is an object type
listings_detail.host_response_rate.unique()
```

* I will convert these values to numeric type.

```python
# removing the % and converting string to numeric and dividing the values to 100 to obtain the rate.
listings_detail['host_response_rate'] = pd.to_numeric(listings_detail['host_response_rate'].str.replace('%',''))/100
```

```python
# checking the unique values for the host acceptance rate since it is an object type
listings_detail['host_acceptance_rate'].unique()
```

* I will convert these values to numeric type.

```python
# removing the % and converting string to numeric and dividing the values to 100 to obtain the rate.

listings_detail['host_acceptance_rate'] = pd.to_numeric(listings_detail['host_acceptance_rate'].str.replace('%',''))/100
```

* the inf values in accommodates_per_bed indicates that some listings have 0 beds wheras they can accommodate guests. Looks like it is a data entry error. I can covert these to NaN and try to do imputation later on. But first I will check how these rows look like and how many rows are like this.

```python
listings_detail[['id','beds','bedrooms','accommodates','beds_per_bedroom','accommodates_per_bed','accommodates_per_bedrooms']].sort_values('accommodates_per_bedrooms',ascending=False).head(20)
```

* I will convert all 0 beds to np.nan and impute based on number of beds per bedrooms or number of people accommodates per bed.

```python
listings_detail['beds'].replace(0, np.nan,inplace=True)
listings_detail['beds_per_bedroom'] = listings_detail['beds']/listings_detail['bedrooms']
listings_detail['accommodates_per_bed'] = listings_detail['accommodates']/listings_detail['beds']
listings_detail['accommodates_per_bedrooms'] = listings_detail['accommodates']/listings_detail['bedrooms']
```

```python
listings_detail[['id','beds','bedrooms','accommodates','beds_per_bedroom','accommodates_per_bed','accommodates_per_bedrooms']].describe()
```

* Median values for both accommodates_per_bedrooms and accommodates_per_bed are 2. I will fill the missing values using accommodates column.

```python
listings_detail['bedrooms'].where(listings_detail['bedrooms'].notnull(),round(listings_detail['accommodates']/2),inplace=True)
listings_detail['beds'].where(listings_detail['beds'].notnull(),round(listings_detail['accommodates']/2),inplace=True)
# if accommodates has 1 then round(accommodates/2) becomes 0. To overcome that I replace 0s with 1.
listings_detail['bedrooms'] = listings_detail['bedrooms'].replace(0,1)
listings_detail['beds'] = listings_detail['beds'].replace(0,1)
```
